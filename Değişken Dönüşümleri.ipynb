```python
import numpy as np
import pandas as pd

V1 = np.array([1,3,6,7,1,9,15])
V2 = np.array([7,5,8,12,2,3,77])
V3 = np.array([1,12,5,6,14,7,31])
df = pd.DataFrame({"V1": V1,
                   "V2": V2,
                   "V3": V3}
                 )

df = df.astype(float)
df
```

## Standardizasyon

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)
```

```python
df.mean()
```

## Normalizasyon

```python
## değerleri 0 ile 1 arasına çeker
preprocessing.normalize(df)
```

## Min Max Dönüşümü 

```python
## Bizim belirlediğimiz aralığa dönüştürür
## Burada 0 ile 2 arasına çektik
preprocessing.minmax_scale(df,feature_range=(0,2))
```

```python
preprocessing.minmax_scale(df,feature_range=(10,20))
```

## Değişken Dönüşümleri

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

### 0 - 1 Dönüşümü

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbe = LabelEncoder()
```

```python
lbe.fit_transform(df["sex"])
```

```python
df["yeni_sex"] = lbe.fit_transform(df["sex"])
```

```python
df
```

### "1 ve Diğerlerini(0)" Dönüşümü

```python
import numpy as np
df["yeni_day"] = np.where(df["day"].str.contains("Sun"),1,0)
df
```
